Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

Load the dataset

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

Necessary preprocessing

```python
data = data.drop(columns=['RowNumber', 'CustomerId', 'Surname'], axis=1)
data.head()
```

```python
data.shape
```

Encoding Categorical variables

```python
#Label Encoding for gender as only 2 categories (male and female)
le_gender = LabelEncoder()
data['Gender'] = le_gender.fit_transform(data['Gender'])
data.head()
```

```python
data.columns
```

```python
data['Geography'].value_counts()
```

```python
#One Hot Encoding for geogrpahy as 3 categories and nominal
ohe_geography = OneHotEncoder(drop='first')
geo_encoded = ohe_geography.fit_transform(data[['Geography']]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=ohe_geography.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
data = pd.concat([data.drop(columns=['Geography']), geo_encoded_df], axis=1)
data.head()
```

```python
data.shape
```

```python
data.info()
```

Necessary EDA(basic descriptive statistics)

```python
exited_counts = data['Exited'].value_counts()
plt.figure(figsize=(6,4))
plt.pie(exited_counts,
        labels=exited_counts.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=['blue', 'red'],
        explode=(0, 0.1),
        shadow=True)
plt.title('Distribution of Exited Customers')
plt.axis('equal')
plt.show()
```

```python
data.corr()
```

```python
data.describe()
```

Splitting features and target

```python
X = data.drop(columns=['Exited'], axis=1)
y = data['Exited']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

Decision Tree

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
clf.score(X_test, y_test)
```

```python
clf.score(X_train, y_train)
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```
