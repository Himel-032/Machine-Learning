# Salary Prediction using Linear Regression

```python
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
df = pd.read_csv("Salary Data.csv", sep=',')
df.head()
```

# EDA

```python
df.isnull().sum()
```

```python
#feature target splitting
x=df[['Experience Years']]
x.head()
```

```python
y=df[['Salary']]
y.head()
```

```python
#scatter plot
plt.scatter(x,y)
plt.scatter(x.mean(), y.mean(), c='r')
plt.xlabel('Experience in Years')
plt.ylabel('Salary')
plt.title('Salary Prediction')
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
m = reg.coef_
m
```

```python
c=reg.intercept_
c
```

```python
m*2.1+c
```

```python
reg.predict([[2.1]])
```

# Cost & Lose

```python
#scatter plot
plt.scatter(x,y)
plt.scatter(x,reg.predict(x),c='b')
plt.scatter(x.mean(), y.mean(), c='r')
plt.plot(x,reg.predict(x))
plt.xlabel('Experience in Years')
plt.ylabel('Salary')
plt.title('Salary Prediction')
```

```python
reg.score(x,y)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse= mean_squared_error(y,reg.predict(x))
mse
```

mae= mean_absolute_error(y, reg.predict(x))
